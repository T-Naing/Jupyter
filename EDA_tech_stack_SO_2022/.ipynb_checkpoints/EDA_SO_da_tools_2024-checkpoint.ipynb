### Helper Function
The survey includes several columns with multiple-option responses. To analyze these choices effectively, we’ll use a helper function that can separate them.

```python
def split_multicolumn(col_series):
    result_df = col_series.to_frame()
    options = []
    # Iterate over the column
    for idx, value  in col_series[col_series.notnull()].items():
        # Break each value into list of options
        for option in value.split(';'):
            # Add the option as a column to result
            if not option in result_df.columns:
                options.append(option)
                result_df[option] = False # Create new column
            # Mark the value in the option column as True
            result_df.at[idx, option] = True 
    return result_df[options]
```

First, we will prepare the data for all developer roles, followed by a subset specifically for data professionals.

```python
cntrs_df=survey_df.Country.value_counts().sort_values(ascending=False).reset_index()
cntrs_df.columns=['Country','Count']
cntrs_df.index = cntrs_df.index + 1
```

```python
cntrs_df['Percentage']=cntrs_df.Count*100/cntrs_df.Count.sum()
# cntrs_df
```

Now that we've prepared the data for all developers, let’s move on to preparing the subset for data professionals.

```python
dcr_cntrs_df= gb_dcr_df.groupby('Country')["dcr"]\
    .sum()\
    .to_frame("Count")\
    .sort_values(by='Count', ascending=False)\
    .reset_index()
```

```python
dcr_total= dcr_cntrs_df.Count.sum()
```

```python
dcr_cntrs_df["Percentage"] = dcr_cntrs_df['Count']/dcr_total*100
```

```python
dcr_cntrs_df.index= dcr_cntrs_df.index+1
```

Now that both datasets are ready, let’s create a side-by-side comparison plot to visualize the differences.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 8))

# whole dataset
axes[0].set_title('Top 10 Countries by Developer Representation')
sns.barplot(y=cntrs_df.head(10).Country, 
            x=cntrs_df.head(10).Percentage, 
            ax=axes[0])

# customize x-axis
x_ticks = np.arange(0, max(cntrs_df.head(10).Percentage) + 5, 5)
x_labels = [f'{tick:.1f}%' for tick in x_ticks]
axes[0].set_xticks(x_ticks)
axes[0].set_xticklabels(x_labels)
axes[0].set_xlabel('Share of Total Developers (%)')

# customize y-axis
y_labels = [f'{idx}. {country}' for idx, country in zip(cntrs_df.head(10).index.astype(str), cntrs_df.head(10).Country)]
axes[0].set_yticks(range(len(y_labels)))
axes[0].set_yticklabels(y_labels)

# customize data labels
for index, value in enumerate(cntrs_df.head(10).Percentage): 
    axes[0].text(value, index, f'{value:.2f}%', va='center')

# Data Professionals dataset
axes[1].set_title('Top 10 Countries by Data Professional Representation')
sns.barplot(y=(dcr_cntrs_df.head(10)).Country, 
            x=(dcr_cntrs_df.head(10)).Percentage,
            ax=axes[1])

x_ticks = np.arange(0, max(dcr_cntrs_df.head(10).Percentage) + 5, 5)
x_labels = [f'{tick:.1f}%' for tick in x_ticks]
axes[1].set_xticks(x_ticks, x_labels)
# axes[1].set_xticklabels(x_labels)
axes[1].set_xlabel('Share of Total Data Professionals (%)')

y_labels = [f'{idx}. {dev_type}' for idx, dev_type in zip(dcr_cntrs_df.head(10).index.astype(str), dcr_cntrs_df.head(10).Country)]
axes[1].set_yticks(range(len(y_labels)))
axes[1].set_yticklabels(y_labels)
# axes[1].set_ylabel

for index, value in enumerate(dcr_cntrs_df.head(10).Percentage): 
    axes[1].text(value, index, f'{value:.2f}%', va='center')

plt.tight_layout()
```

##### Top 10 Programming Languages Among Data Professionals in Selected Countries
Now, let’s explore how programming language usage varies across the selected countries—Thailand, Vietnam, Singapore, Japan, the United Kingdom, and the United States.

This regional breakdown reveals differences in tool preferences, educational exposure, and industry demands, offering a more nuanced view of how data professionals approach their work across global tech hubs.

```python
fig, axs = plt.subplots(3, 2, figsize=(18, 15))

# Flatten the array of axes for easier iteration
axs = axs.flatten()

for i, country in enumerate(ord_sel_cntrs):
    # print(i, country)
    lang_worked_cou_df = split_multicolumn(gb_dcr_df[gb_dcr_df.Country==country].LanguageHaveWorkedWith)
    lang_worked_cou_pct = (lang_worked_cou_df.mean().sort_values(ascending=False) * 100).head(10).reset_index()
    lang_worked_cou_pct.columns=['Languages Developers Have Worked With','Developer Percentage']
    lang_worked_cou_pct.index += 1
    # print(languages_worked_percentages)
    
    sns.barplot(ax=axs[i], 
                x='Developer Percentage', 
                y='Languages Developers Have Worked With', 
                data=lang_worked_cou_pct)
    axs[i].set_title(f'Top 10 Programming Languages Used by Data Professionals in {country}')
    axs[i].set_xlabel(f'Share of Data Professionals (%)')

    for index, value in enumerate(lang_worked_cou_pct['Developer Percentage']): 
        axs[i].text(value, index, f'{value:.2f}%', va='center')

    y_labels = [f'{i}. {value}' for i, value in zip(
        lang_worked_cou_pct.index.astype(str), 
        lang_worked_cou_pct['Languages Developers Have Worked With'])]
    axs[i].set_yticks(range(len(y_labels)))
    axs[i].set_yticklabels(y_labels)
    axs[i].set_ylabel('Programming Language')

    # Customizing x-ticks with percentage symbols
    x_ticks = range(0, int(max(lang_worked_cou_pct['Developer Percentage'])) + 10, 10)
    x_labels = [f'{tick}%' for tick in x_ticks]
    axs[i].set_xticks(x_ticks)
    axs[i].set_xticklabels(x_labels)
    
plt.tight_layout()
# plt.savefig('./Comparative-Study-of-Myanmar-Coding-Community/10 Most Used Programming Languages in.png')
plt.show();
```

> Python stands out as the most dominant programming language among data professionals, both globally and across all the selected countries. Interestingly, R does not appear among the most commonly used languages.
>
>In Singapore, 78.26% of data professionals report using Python—the highest rate among all countries analyzed. However, only 56.52% have experience with SQL, reflecting a notably lower usage compared to most other regions, except Vietnam, where just 51.85% of data professionals report using SQL.
>
>In contrast, in the United States, usage of Python and SQL is nearly balanced, with 70.63% of respondents reporting experience with Python and 69.07% with SQL. This reflects a strong dual reliance on both languages within the U.S. data ecosystem.

### Education Level
Since programming language usage shows little variation by country, we aim to examine whether language preferences differ by education level.

```python
lang_ed_df = lang_worked_gbl_df.copy()
lang_ed_df ['EdLevel'] = gb_dcr_df['EdLevel'] 
```

```python
lang_ed_df = (lang_ed_df.groupby("EdLevel").mean()*100)
# lang_ed.head()
```

```python
lang_ed_df = (
    lang_ed_df  # Make 'EdLevel' the row index
      .apply(lambda row: row.nlargest(3), axis=1)  # Top 3 values for each row
)
lang_ed_df
```

```python
# df_top3.stack()
```

```python
# Convert to long format for visualization
lang_ed_df = lang_ed_df.stack().reset_index()
lang_ed_df.columns = ['EdLevel', 'Language', 'Percentage']
lang_ed_df.head()
```

```python
custom_order = [
    'Primary/elementary school',
    'Secondary school',
    'Associate degree',
    'Some college',
    'Bachelor’s degree',
    'Master\'s degree',
    'Professional degree',
    'Other doctoral degree',
    'Something else'
]

lang_ed_df['EdLevel'] = pd.Categorical(
    lang_ed_df['EdLevel'],
    categories=custom_order,
    ordered=True
)

# Sort EdLevel ascending, Percentage descending
lang_ed_df = lang_ed_df.sort_values(
    by=['EdLevel','Language', 'Percentage'],
    ascending=[True, True, False]
).reset_index(drop=True)

lang_ed_df.head()
```

```python
plt.title('Programming Language Usage by Education Level among Data Professionals',
         pad= 20)

sns.barplot(x='EdLevel', 
            y='Percentage', 
            hue='Language', 
            data=lang_ed_df,
            dodge=True ,
            order=custom_order
           );

plt.xticks(rotation=90)
plt.xlabel('Education Level')

y_ticks = range(0, int(lang_ed_df['Percentage'].max()) + 10, 10)
y_labels = [f'{tick}%' for tick in y_ticks]
plt.yticks(y_ticks, y_labels)
plt.ylabel('Share of Data Professionals (%)')

plt.legend(
    title='Language',
    bbox_to_anchor=(1.05, 1),
    loc='upper left',
    borderaxespad=0
);
```

>According to the survey, Python is primarily used by data professionals who hold some college education, Bachelor's, Master's, professional (e.g., JD, MD), or other doctoral (e.g., Ph.D., Ed.D.) degrees.
>
>Among these, Bachelor's degree holders use Python and SQL at nearly the same rate, while Python usage is highest among Master's (79.08%) and doctoral (83.1%) degree holders.
>
>Conversely, those with primary, secondary, or associate degrees rarely use Python. In contrast, SQL is utilized by data professionals across all education levels.
